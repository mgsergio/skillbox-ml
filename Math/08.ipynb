# Задание 2

```
observations = [
  [1945, 560, 4.3],
  [1495, 340, 4.9],
  [1570, 343, 5.2]
]
X = [1520, 431]
```

```
a = sp.symbols("a:3")
a
```

```
terms = [(x[0] * a[0] + x[1] * a[1] + a[2] - x[2])**2 for x in observations]
terms
```

```
from functools import reduce
mse = reduce(
    lambda f, term: f + term,
    terms
) / len(terms)
mse
```

```
d_mse = [sp.diff(mse, a[i]) for i in range(len(a))]
d_mse
```

```
sln = sp.linsolve(d_mse, a)
sln
```

```
sln = next(iter(sln))
sln
```

```
[d_mse[i].subs(dict(zip(a, sln))) for i in range(len(sln))]
```

Поскольку $mse > 0 \space\forall a = (a_1, a_2)$, то нули производной — либо миним $f$ либо точка перегиба. Не знаю, как доказать, что в данном случае мы получили именно экстремум, поэтому просто поверю в это.


```
f = reduce(
    lambda f, term: f + term,
    [a*x for x, a in zip(sp.symbols("x:2"), sln[:-1])]
) + sln[-1]
f
```

```
sp.plotting.plot3d(f);
```
