```
import sympy as sp
```

```
X = sp.Matrix([
  [1, 2],
  [5, 3],
  [2, 4],
  [3, 7]
])
X
```

```
B = sp.Matrix([5, 6, 10, 8])
B
```

```
psydo_inverse = (X.T * X) ** -1 * X.T
psydo_inverse
```

```
A = psydo_inverse * B
A
```

```
mse = X * A - B
mse = mse.T * mse / mse.rows
float(mse[0])
```

```
A.applyfunc(float)
```

Градиентный спуск в задании 9 дал $mse = 6.35$ в точке $(0.68, 1.17)$. Метот псевдообратных матриц улучшил результат $mse$ до $6.24$ в точке $(0.49, 1.30)$.
