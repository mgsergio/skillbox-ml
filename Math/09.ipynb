```
from itertools import chain
from typing import List, TypeVar


T = TypeVar('T')


def check_rows(rs: List[List[T]]) -> None:
  lengths = [len(r) for r in rs]
  if not all([l == lengths[0] for l in lengths]):
    raise ValueError("All rows mast have the same number of columns.")


def print_matrix(m: List[List[T]]) -> None:
  check_rows(m)
  max_length = max([
    len(str(item)) for item in chain.from_iterable(m)
  ])
  fmt = f"{{0: ={max_length}d}}"
  print("[")

  for r in m:
    print(f"  [{', '.join([fmt.format(e) for e in r])}]")

  print("]")


def transpose(m: List[List[T]]) -> List[List[T]]:
  check_rows(m)

  return [
    [m[r][c] for r in range(len(m))]
    for c in range(len(m[0]))
  ]


def transpose2(m: List[List[T]]) -> List[List[T]]:
  check_rows(m)
  return list(zip(*m))
```

```
A = [
  [ 2,  1,  7,  4],
  [ 5,  6,  7,  3],
  [ 9,  8,  2, 12],
  [11, 14, 15, 15],
]

print_matrix(A)
```

```
A_t = transpose(A)
print_matrix(A_t)
```

```
B = [
  [3, 7, 8, 3,  6],
  [2, 5, 9, 4, 13]
]
print_matrix(B)
```

```
B_t = transpose2(B)
print_matrix(B_t)
```

# Задание 3

```
A = sp.Matrix(sp.symbols("a1:3"))
A
```

```
X = sp.Matrix([
  [1, 2],
  [5, 3],
  [2, 4],
  [3, 7]
])
X
```

```
Y = sp.Matrix([5, 6, 10, 8])
```

```
mse = (X * A - Y)
mse = (mse.T * mse / len(mse))[0]
mse
```

```
def grad_decent(f, start, vars, step, eps):
  for_subs = lambda x: dict(zip(vars, x))
  d_f = sp.Matrix([sp.diff(f, v) for v in vars])

  X = sp.Matrix(start)
  path = [list(X)]
  while f.subs(for_subs(X)) > eps:
    X -= step * d_f.subs(for_subs(X))
    path.append(list(X))

  return dict(
      argmin=path[-1], 
      min=f.subs(for_subs(X)), 
      path=path
  )
```

```
grad_decent(mse, [0, 0], list(A), 0.01, 6.36)
```
