### Задание 8

Отсортировать строки в датафрейме в соответствии со значениями `price` и `income` в порядке возрастания. Можно ли сказать, что люди с более низким доходом и выбравшие более дешёвый тип рыбалки, в целом, предпочитают один тип рыбалки, а люди с более высоким доходом и более дорогой рыбалкой – другой? Ответ записать в виде текстовой ячейки или в виде комментария.

```python
### YOUR CODE HERE ###
preferences = dat[['mode','price', 'income']].sort_values(['price', 'income'])
preferences
```

Просто по выхлопу сортировки с уверенностью сказать о предпочтненни нельзя, так как в выводе видно только 6 первых и последних записей, которые могли вполне слуайно распределиться таким образом, что будет предполагаться связь типа рыбалки и дохода. Я бы посмотрел еще на распределение простые метрики по группам, возможно есть очевидный способ отделить типы рыбалки по цене и доходу.  

```python
preferences.groupby('mode').agg(['min', 'max', 'mean', 'var'])
```

Видно, что в стреднем charter дороже других типов рыблаки, дале идет boat, но уж больно большой разброс видно по var и в цене и в доходе. Еще видно, что люди с самым низким доходом также могут позволить себе рыбалку charter.
Думаю, стоит посмотреть на респределение типов рыбкалки по доходу.

```python
mode_income = pd.DataFrame({'income': pd.cut(dat['income'], 7), 'mode': dat['mode']})
mode_income_pivot = (
    mode_income
    .assign(ones=1)
    .pivot_table(
        index='income', columns='mode', 
        values='ones', aggfunc='count'
    )
)
mode_income_pivot
```

```python
mode_income_pivot.div(mode_income_pivot.sum(axis=1), axis=0) * 100
```

Видно, что хорошо обеспеченные люди тяготеют к рыбалке с лодки. Рыбалку с пляжа (или с берега) недолюбливают все вне зависимости от достатка.
