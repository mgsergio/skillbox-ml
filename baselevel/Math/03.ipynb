### Задание 2. Проведите исследование полученных функций. 

1. $5x + 1.5$
2. $2e^{0.2x}-1$
3. $3\sqrt x - 15$
4. $2\log(x) + 5$
5. $-5x^2 - 10x + 100$




```python
from IPython.display import display, Markdown
from sympy import exp, init_printing, latex, log, plot, solve, sqrt, symbols
from sympy.plotting import plot

init_printing()

x = symbols("x", real=True)
```

```python
def analyze(f):
    display(f)
    plot(f)
    display(        
        ["zeors:", solve(f)], 
        [f < 0, "on", solve(f < 0)], 
        [f > 0, "on", solve(f > 0)],
        ["f is even:", f == f.subs(x, -x)],
        ["f is odd:", f == -f.subs(x, -x)]
    )
```

```python
analyze(5*x + 1.5)
```

возрастает на (-inf; +inf)

```python
analyze(2*exp(0.2*x) - 1)
```

возрастает на (-inf; +inf)

```python
analyze(3*sqrt(x) - 15)
```

возрастает на [0; +inf)

```python
analyze(2*log(x) + 5)
```

возрастает на (0; +inf)

```python

analyze(-5*x**2 - 10*x + 100)
```
